```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
gdf = gpd.read_file('data_file.geojson')
gdf.sort_values(by=['ED_ID'])
```

```python
plt.figure()

ax = plt.subplot()

ax.set_title('Population Density ($\\ln\\frac{P}{A}$)')
ax.set_xticks([])
ax.set_yticks([])

gdf.plot(ax = ax, column='density', legend=True,  cmap='inferno')
```
